```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# 1. 读取Excel文件
file_path = 'DATA.xlsx'
df = pd.read_excel(file_path)

# 2. 提取水化学指标部分（从第2列到第26列）
chem_data = df.iloc[:, 1:26]

# 3. 数据标准化
scaler = StandardScaler()
chem_data_scaled = scaler.fit_transform(chem_data)

# 4. PCA降维（只保留前5个主成分）
n_components = 5
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(chem_data_scaled)

# 5. 输出降维后的结果
principal_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(n_components)])
print("PCA降维后的数据:")
print(principal_df.head())

# 6. 输出主成分方差解释比例
print("每个主成分的解释方差比：")
for i, ratio in enumerate(pca.explained_variance_ratio_):
    print(f"PC{i+1}: {ratio:.4f}")

# 7. 输出每个主成分的因子载荷矩阵（原始变量在主成分上的贡献）
loadings = pd.DataFrame(pca.components_.T,
                        columns=[f'PC{i+1}' for i in range(n_components)],
                        index=chem_data.columns)
print("\n主成分因子载荷矩阵（每个主成分包含的原始变量权重）:")
print(loadings)

# 可选：将主成分与原始因子的载荷保存为Excel
loadings.to_excel("PCA_Loadings.xlsx")
principal_df.to_excel("Jiyuan_PCA_5Components.xlsx", index=False)
# 提取每个主成分中贡献最大的变量（默认取前三个）
top_n = 3
important_factors = {}

for pc in loadings.columns:
    top_vars = loadings[pc].abs().sort_values(ascending=False).head(top_n).index.tolist()
    important_factors[pc] = top_vars

# 输出每个主成分对应的主要因子
print("\n每个主成分的主要贡献因子（按绝对载荷值排序）：")
for pc, factors in important_factors.items():
    print(f"{pc} 主要受以下变量影响：{', '.join(factors)}")
```
